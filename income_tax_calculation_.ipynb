```python
# Code by Ayush Agarwal IIT BHU ECE24

# Code to calculate tax amount as per new tax regime and hence find effective tax percentage

import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# input variable x would be the income

def calculate_income_tax(x):
  if(x<300000):
    return 0
  elif(x<600000 and x>300000):
    return (0+(0.05*(x-300000)))
  elif(x<900000 and x>600000):
    return (0 + (0.05*(600000-300000)) + (0.10*(x-600000)))
  elif(x<1200000 and x>900000):
    return (0 + (0.05*(600000-300000)) + (0.10*(900000-600000)) + (0.15*(x-900000)))
  elif(x<1500000 and x>1200000):
    return (0 + (0.05*(600000-300000)) + (0.10*(900000-600000)) + (0.15*(1200000-900000)) + (0.20*(x-1200000)))
  elif(x>1500000):
    return (0 + (0.05*(600000-300000)) + (0.10*(900000-600000)) + (0.15*(1200000-900000)) + (0.20*(1500000-1200000)) + (0.30*(x-1500000)))

# print(calculate_income_tax(1800000))

plt.figure(figsize=(7, 5))

# not starting from 0 since it was causing division by 0 error
x1 = np.linspace(1,2000000,1000)
y1 = [calculate_income_tax(a) for a in x1]
plt.xlabel("income value")
plt.ylabel("amount of tax to be paid")
plt.title("INCOME TAX CALCULATOR")
plt.ticklabel_format(style='plain')  # Display x-values in full
plt.grid(True)
plt.axvline(x=0, color='red', linestyle='--')
plt.axvline(x=300000, color='red', linestyle='--')
plt.axvline(x=600000, color='red', linestyle='--')
plt.axvline(x=900000, color='red', linestyle='--')
plt.axvline(x=1200000, color='red', linestyle='--')
plt.axvline(x=1500000, color='red', linestyle='--')

plt.plot(x1,y1)
plt.show()

plt.figure(figsize=(7, 5))

y2 = [(((calculate_income_tax(a)*100))/a) for a in x1]
plt.ticklabel_format(style='plain')  # Display x-values in full
plt.grid(True)
plt.axvline(x=0, color='red', linestyle='--')
plt.axvline(x=300000, color='red', linestyle='--')
plt.axvline(x=600000, color='red', linestyle='--')
plt.axvline(x=900000, color='red', linestyle='--')
plt.axvline(x=1200000, color='red', linestyle='--')
plt.axvline(x=1500000, color='red', linestyle='--')
plt.plot(x1,y2)
plt.xlabel("income value")
plt.ylabel("effective tax %")
plt.title("EFFECTIVE TAX PERCENTAGE CALCULATOR")
plt.show()
```
