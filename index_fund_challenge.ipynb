```python
cagr_nifty = []

for i in range(0,len(nifty_values)-12):
  cagr_nifty.append(  (  ((nifty_values[i+12]-nifty_values[i])/nifty_values[i])  )  *100    )

print(cagr_nifty)
```

```python
# investing for any 1 year in nifty in between 2014 start to 2022 end , can it beat inflation ?
did_it_beat = cagr_nifty>inflation_rates_yoy
print(did_it_beat)
print(np.count_nonzero(did_it_beat))
print(np.count_nonzero(did_it_beat)/len(did_it_beat))
print((len(did_it_beat)-np.count_nonzero(did_it_beat)))
print((len(did_it_beat)-np.count_nonzero(did_it_beat))/len(did_it_beat))

# 65% times it beats inflation , 35% it dosent
# So I may or may not get fucked

# when we invest for 1 yr at any time , the inflation adjusted CAGR returns would be -
yr_1_inf_adj_rtrn = cagr_nifty-inflation_rates_yoy

sns.kdeplot(yr_1_inf_adj_rtrn)
plt.axvline(x=0, color='red', linestyle='--')
plt.axvline(x=np.mean(yr_1_inf_adj_rtrn), color='red', linestyle='--')
plt.axvline(x=np.mean(yr_1_inf_adj_rtrn)+np.std(yr_1_inf_adj_rtrn), color='red', linestyle='--')
plt.axvline(x=np.mean(yr_1_inf_adj_rtrn)-np.std(yr_1_inf_adj_rtrn), color='red', linestyle='--')
plt.plot()

# https://seaborn.pydata.org/generated/seaborn.kdeplot.html
```

```python
print(cagr_nifty)
print(inflation_rates_yoy)

# matrix which will hold arrays of different time period returns M
M = []
temp = []
M.append(yr_1_inf_adj_rtrn)

# print(M)
# print()
# print()

# temp.append(1)
# M.append(temp)
# print(M)

# j is starting value iterates over all
# i is number of years
# k is for calculating each individual element so looping term

no_yrs_strategy = 15
for i in range(2,no_yrs_strategy):
  # print(i)
  temp = []
  for j in range(0,len(nifty_values)-(12*i)):
    # print(j)
    new_cagr = 1
    new_inf = 1
    for k in range(0,i):
      new_cagr = new_cagr*(1+(cagr_nifty[j+(12*k)]/100)) #1.08
      new_inf = new_inf*(1-(inflation_rates_yoy[j+(12*k)]/100)) #0.97
    temp.append(100*((new_cagr-1)-(1-new_inf)))
  M.append(temp)

# print(M)

max_returns = []
min_returns = []
avg_returns = []

for i in range(0,len(M)):
  sns.kdeplot(M[i])
  plt.axvline(x=0, color='red', linestyle='--')
  plt.axvline(x=np.mean(M[i]), color='red', linestyle='--')
  plt.axvline(x=np.mean(M[i])+np.std(M[i]), color='red', linestyle='--')
  plt.axvline(x=np.mean(M[i])-np.std(M[i]), color='red', linestyle='--')
  plt.xlim(-150,300)
  plt.plot()
  plt.savefig('yrs_'+str(i)+'nifty'+'.jpg')
  plt.close()
  print(len(M[i]))
  max_returns.append(max(M[i]))
  min_returns.append(min(M[i]))
  avg_returns.append(sum(M[i])/len(M[i]))

print(max_returns)
print(min_returns)
print(avg_returns)
print(len(M))
# x1 = np.linspace(0,8,1)
plt.plot(max_returns)
plt.plot(avg_returns)
plt.plot(min_returns)
plt.axhline(y=0, color='red', linestyle='--')
plt.show()
```
